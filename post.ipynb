```python
import pandas as pd
```

```python
cf11_tienda = pd.read_excel('output/cierres_f11/tienda/221104-1438-cf11_tienda_21-output.xlsx', dtype= str, sheet_name = 'DB')
```

```python
dv = pd.read_csv('output/bases/221006-1551-221004_dv-output.csv', sep=';', dtype= str)
```

```python
dv['F11'] = dv['NOTAS'].str.extract(r'([1][1]\d{7,})')
```

```python
dv = dv[['F11', 'SKU ORIGINAL', 'CANT. UND GENERADAS']]
```

```python
dv.drop_duplicates(['F11', 'SKU ORIGINAL'], inplace = True)
```

```python
dv.reset_index(drop=True, inplace=True)
```

```python
cf11_tienda
```

```python
cf11_dv = cf11_tienda.merge(dv, how='left', left_on = ['folio_servicio_tecnico', 'sku_(numero_de_producto)'], right_on = ['F11', 'SKU ORIGINAL'])
```

```python
cf11_dv['Comentario GCO'].unique()
```

```python
cf11_dv['motivo_cierre'].unique()
```

```python
cf11_dv.loc[(cf11_dv['motivo_cierre'] == 'dv') & (cf11_dv['F11'].isna()), ['Comentario GCO', 'checked']] = ['No coincide (F11|DV)', 'y']
```

```python
cf11_dv.loc[cf11_dv['F11'].notna() & (cf11_dv['cantidad_f11'] == cf11_dv['CANT. UND GENERADAS']), ['motivo_cierre','Comentario GCO', 'checked']] = ['dv','Coincidencia Exacta (F11|DV)+UPC+QTY', 'y']
cf11_dv.loc[cf11_dv['F11'].notna() & (cf11_dv['cantidad_f11'] != cf11_dv['CANT. UND GENERADAS']), ['motivo_cierre','Comentario GCO', 'checked']] = ['dv','Coincidencia Exacta (F11|DV) QTY diferente', 'y']
```

```python
cf11_dv.to_excel(f'output/cierres_f11/tienda/{dt}_post.xlsx', index = False)
```
